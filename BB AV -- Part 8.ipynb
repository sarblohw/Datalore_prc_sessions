```python
# Using recursion to define a functione to calculate factorial of a number

def fact(num):
    if num == 0:
        return 1
    else:
        return num * fact(num-1)
```

```python
def factorial_self(n):
    if n == 0:
        return 1
    else:
        return n * factorial_self(n-1)
```

```python
factorial_self(10)
```
